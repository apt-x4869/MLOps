```python
from scipy import misc
from scipy.signal import convolve2d
import numpy as np
```

```python
img = misc.ascent()
```

```python
img.shape
```

```python
kernel = np.array([[-1,-1,-1],[-1,4,-1],[-1,-1,-1]])
```

```python
c = convolve2d(img,kernel)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(c, cmap='gray')
```
