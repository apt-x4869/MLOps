```python
import plotly.graph_objs as go
```

```python
import plotly.offline as pyo
```

```python
datax = np.random.randint(1,101,100)
```

```python
datay = np.random.randint(1,101,100)
```

```python
graph_data = [ go.Scatter(x=datax, y=datay, mode='markers'), go.Bar(), go.Line() ]
```

```python
layout = go.Layout(title='my graph title')
```

```python
fig = go.Figure(data=graph_data, layout=layout)
```

```python
pyo.plot(fig, filename='mygraph.html')
```
