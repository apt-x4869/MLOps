```python
import cv2
import numpy as np
```

```python
photo = cv2.imread('cv.jpeg')
```

```python
cv2.imshow('cv' ,photo+255)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
x = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)
x[10]=200
```

```python
cv2.imshow('cv' ,x)
cv2.waitKey(9000)
cv2.destroyAllWindows()
```
