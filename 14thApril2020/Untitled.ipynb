```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
status, photo = cap.read()
cv2.imshow('hi',photo)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
photo = cv2.rectangle(photo, (200,50),(400,300),[0,255,0], 3)
```

```python
cv2.imshow('hi',photo)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap.release()
```
