```python
# Down load IP webcam to send video data from there
# our python code will read the video data and can display it

import cv2 
import numpy as np
url = 'http://192.168.31.139:8080/video'
cap = cv2.VideoCapture(url)
while(True):
    ret, frame = cap.read()
    if frame is not None:
        cv2.imshow('frame',frame)
    q = cv2.waitKey(10)
    if q == ord("q"):
        break
cv2.destroyAllWindows()
cap.release()
```
