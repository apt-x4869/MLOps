```python
import cv2
```

```python
# to connect webcam 0-->default 1-->another webcam ....
cap = cv2.VideoCapture(0)
```

```python
status, photo = cap.read()
```

```python
cap.release()
```

```python
cv2.imshow('hi',photo)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
photo.shape
```

```python
croped = photo[100:400, 200:550]
```

```python
cv2.imshow('hi2', croped)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
croped.shape
```

```python
croped_resize = cv2.resize(croped, (100,100))
```

```python
croped_resize.shape
```

```python
photo[0:100,0:100,:] = croped_resize
```

```python
cv2.imshow('hi',photo)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
