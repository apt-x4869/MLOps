```python
import cv2
```

```python
photo = cv2.imread('photo.jpg')
```

```python
type(photo)
```

```python
photo.shape
```

```python
photo[: , : , :]
```

```python
# Removing Red Colour
a2 = photo
a2[: , : , 2] = 0
```

```python
cv2.imshow('hi', photo)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
# Colour Amplification
a3 = photo
a3[: , : , 0] += 100
```

```python
cv2.imshow('hi', a3)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
# split() function
B, G, R = cv2.split(photo)
# zeores is of dtype float
zeros = np.zeros((288 , 552))
z = zeros.astype('uint8')
```

```python
# merge() function
a4 = cv2.merge([B, G, z])
```

```python
cv2.imshow('hi', a4)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
