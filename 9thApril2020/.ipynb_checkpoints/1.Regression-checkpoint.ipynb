```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('db.csv')
```

```python
dataset
```

```python
y = dataset['marks']
```

```python
y
```

```python
x = dataset['duration']
```

```python
type(x)
```

```python
X = x.values.reshape(4,1)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X , y)
```

```python
model.predict([[9]])
```

```python
model.coef_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```
