```python
import pandas as pd
```

```python
dataset = pd.read_csv('weight-height.csv')
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
y = dataset['Weight']
```

```python
x = dataset['Height']
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
# by default : Linear activation functions
# units : output
# input shope : input feature
# dense : how many hidden layer
model.add(Dense(units=1 , input_shape=(1,) , ))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.8))
```

```python
model.fit(x,y, epochs=20)
```

```python
w,b = model.get_weights()
```
